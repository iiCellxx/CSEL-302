# 1. Import Libraries

```python
import pandas as pd #Data Manipulation
import numpy as np #Numerical Operation
import matplotlib.pyplot as plt #Data Visualization
from sklearn.linear_model import LinearRegression #Data Modeling
```

# 2. Create a DataFrame

```python
df = pd.read_csv("students_grade.csv")

print(df.head())
```

# 3. Scattering The Data From CSV File

```python
df = pd.read_csv("students_grade.csv")

# Select columns for scatter plot
x_column = "Previous_Exam_Scores"
y_column = "Grades"

# Create the scatter plot
plt.scatter(df[x_column], df[y_column], label="Data points", color="blue")
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title(f"Scatter Plot: {x_column} vs. {y_column}")
plt.grid(True)
plt.legend()
plt.show()
```

# 5. Preparing Data for Modeling

```python
x_column = df[['Previous_Exam_Scores']] #Independent Variable
y_column = df[['Grades']] #Dependent Variable
```

# 6. Fitting the Linear Regression Model

```python
model = LinearRegression()

model.fit(x_column,y_column)
```

# 7. Getting Model Coefficients

```python
slope = model.coef_[0]

intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)
```

# 8. Visualizing The Linear Regression Line

```python
plt.scatter(x_column,y_column)
plt.plot(x_column,model.predict(x_column), color='red')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title('Linear Regression')
plt.show()
```
