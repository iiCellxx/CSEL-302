1.Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

2.Import Dataset


```python
data = pd.read_csv('sales_data_2.csv')
data.head()
```

3.Data Loading and Initial Analysis

```python
Stats = data.describe()
```

4. Data Manipulation

```python
product_revenue = data.groupby('Product')['Revenue'].sum()
highest_revenue_product = product_revenue.idxmax()
average_quantity_per_day = data['Quantity Sold'].mean()
```

5. Data Visualisation

```python
product_revenue.plot (kind = 'bar')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.title('Revenue by Product')
plt.xticks(rotation=45)
plt.show()
```

6. Analysis Report


```python
report = """
Analysis report

1. Data Loading and Initial Analysis:
    Basis statistic of the dataset:
    {}
2. Data Manipulation:
    Total revenue for each product:
    {}
    Product with the highest revenue:{}
    Average quantity sold per day: {:.2f}
3. Data Visualization:
    Bar char displaying revenue for each product is attached.
4. Insights:
    The product {} generates the highest revenue.
    On average, {:.2f} units are sold per day.

""". format(Stats,product_revenue,highest_revenue_product,average_quantity_per_day,
            highest_revenue_product,average_quantity_per_day)

print(report)

```

7. Saving the Report to a File


```python
with open('Report.txt','w') as f:
     f.write(report)
```
